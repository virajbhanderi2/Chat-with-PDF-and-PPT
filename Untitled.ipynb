<h2>Data quality analysis</h2>

```python
reader = PdfReader("AI_ML_NOTES.pdf")
print("Number of pages = ",len(reader.pages))
```

```python
# inspect individual page
page2 = reader.pages[2]
print(page2.extract_text()[:700])
```

```python
# check empty page
# enumerate() is a built-in Python function that lets you loop with an index (counter) and a value at the same time.
empty_page =[]

for i,page in enumerate(reader.pages):
    if not page.extract_text():
        empty_page.append(i)

print("Empty Pages : ",empty_page)
```

```python
# Full text Extraction
text = ""

for page in reader.pages:
    if page.extract_text():
        text+=page.extract_text()

len(text)
```

```python
# get pdf metadata
metadata = reader.metadata
print("Title:", metadata.title)
print("Author:", metadata.author)
print("Creator:", metadata.creator)
print("Producer:", metadata.producer)
```

```python
# Extract text for NLP

full_text = ""

for page in reader.pages:
    text = page.extract_text()
    if text:
        full_text += text + "\n"

print(full_text[:1000])  # preview
```

<h2>HuggingFace Embedding</h2>

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
# Embbed some sentence
vector = embeddings.embed_query("What is AI?")
len(vector)
```

```python
v1 = embeddings.embed_query("CNN is used for image processing")
v2 = embeddings.embed_query("Convolutional neural networks process images")
v3 = embeddings.embed_query("Football is a sport")
```

```python
# np.linalg.norm(a) - This calculates the length of a vector.
# np.dot - Multiplies matching numbers and adds them

import numpy as np

def cosine_similarity(a,b):
    return np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b))

print("v1 vs v2:", cosine_similarity(v1, v2))
print("v1 vs v3:", cosine_similarity(v1, v3))
```

```python
from langchain_community.vectorstores import FAISS
```

```python
# Create vector database using FAISS (avoids chromadb / pydantic issues on Python 3.14)
vector_db = FAISS.from_texts(
    texts=chunks,          # your chunked PDF text
    embedding=embeddings
)
```

```python
# Use FAISS vector store instead of Chroma to avoid chromadb / pydantic issues on Python 3.14
from langchain_community.vectorstores import FAISS

db = FAISS.from_texts(texts=chunks_overlap, embedding=embeddings)
```

```python
# Search similiar text
results = db.similarity_search("what is ml",k=3)
results
```

```python
for i, r in enumerate(results):
    print(f"\nResult {i+1}:\n", r.page_content[:-1])
```

<h2>Ollama LLM</h2>

```python
from langchain_ollama.llms import OllamaLLM

def load_llm(model_name="llama3.2"):
    return OllamaLLM(model=model_name)

llm = load_llm()

print(llm.invoke("What is artificial intelligence?"))
```

<h2>Retriever & RetrievalQA (RAG)</h2>

```python
# Build retriever from the FAISS vector database
retriever = vector_db.as_retriever(search_kwargs={"k": 5})
retriever
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_template("""
You are an honest assistant.
You will answer questions based only on the course PDF content.
If you don't know the answer, just say you don't know. Don't try to make up an answer.

Context:
{context}

Question:
{question}
""")
```

```python
from langchain_core.runnables import RunnablePassthrough

def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)

rag_chain = (
    {
        "context": retriever | format_docs,
        "question": RunnablePassthrough()
    }
    | prompt
    | llm
)
```

```python
# Example RAG query over AI_ML_NOTES.pdf
rag_chain.invoke("What is machine learning and list its main types?")
```

```python
rag_chain.invoke("Explain supervised vs unsupervised learning.")
```
